### Average v Median House Price Values
The obvious takeaway from the boxplot above is that Dublin house prices contain many outlier values. It's important to notice the difference between median and average values in a dataset with as many outliers as this one.

The **average** house price in Dublin in 2016 was €403k. That figure is calculated by adding up all the house prices and dividing that total by the number of houses sold. And that's fine, except that every time a house price appears that is considerably bigger than the majority, it tends to drag up the average. Consider two housing estates of five houses each: houses in estate A sell at €200k, €400k, €300k, €300k and €300k. Estate B has four houses for sale at €100k and a fifth for sale at €1,100k. Both estates have average house prices of €300k, but this give an unreliable impression of house-price parity that doesn't exist in real life.

The **median** house price in Dublin in 2016 was €310k. The median figure is calculated by listing all the prices of houses sold in Dublin from highest to lowest, and then chosing the middle one. So if we say the median price was €310k, that means half the houses in Dublin sold for less than €310k in 2016, and the other half sold for more. It's a more robust measure of the mid-point in the dataset, because one super-high value isn't going to distort it to the same extent as a super-high value can distort an average. In our estate example above, the median house price in Estate A is €300k while the median house price in Estate B is €100k - a much more accurate reflection of the overall picture.

### Average v Median Prices by EirKeyName
With all that in mind, let's look at average and median prices by area in Dublin in 2016.

```python
pcs = houses.groupby('eirKeyName')
house_price_summary_list = []
house_price_summary_list.append(["eirKeyName",
                                "Houses Sold",
                                "Average Price",
                                "Median Price",
                                "Difference between Average and Median Price"])
for postCode, details in pcs:
    house_price_summary_list.append([postCode,
                                    details.Price.count(),
                                    details.Price.mean(),
                                    details.Price.median(),
                                    details.Price.mean() - details.Price.median()])

summary = pd.DataFrame(house_price_summary_list[1:], columns = house_price_summary_list[0])
summary.sort_values('Houses Sold', ascending=False)
```

#### Observation
The area in which most houses were sold in Dublin was Clonee/Blanchardstown/Castleknock - Dublin 15, and environs. The next closest is Kiliney, and then Dublin 12. Why was the north-west of the city the best place to buy a house in 2016? The data can't tell us I'm afraid.

#### Greatest Difference Between Average and Median House Prices

```python
summary.sort_values("Difference between Average and Median Price", ascending=False).head(7)
```

#### Observation
It is hardly surprising that the post code with the greatest difference between the average and median house prices is Dublin 4. When things go high in that part of the city, they go very high indeed. Dublin 6 in second place has less than half as much of a difference between the median and average prices.

#### Least Difference Between Average and Median House Prices

```python
summary.sort_values("Difference between Average and Median Price").head(7)
```

#### Observation
It is statistically remarkable that for Dublin 20 and Dublin 10, Chapelizod/Palmerstown and Ballyfermot, respectively, the average value of a house is less than the median value. No houses are free; there is always a basement price and everything is higher than that, making an average almost inevitably higher than a mean value for any spread of values over thirty. For house prices in Dublin 20 and Dublin 10 to drag the average down is a depressing situation for people living in those houses.

### Houses Sold by eirKeyName
Let's look at a boxplot of house prices broken down by eirKey. We'll only include those postal areas for which we have twenty or more data points.

```python
houses.eirKeyName.value_counts()
```

```python
houses_list = list(houses.eirKeyName.value_counts().keys())
fig, ax = plt.subplots(1,1, figsize = (10, 6))
eirKeyBoxplotData = houses[houses['eirKeyName'].isin(houses_list[:-3])]
eirKeyBoxplotData.boxplot('Price', 'eirKeyName', rot=45, ax=ax)
```

### Removing More Outliers
We see outlier house prices here that are so large they are crushing the distribution visibility of the others. Therefore, we'll cap our house prices according to the standard 1.5 times the 75th percentile, and look again.

```python
eirKeyBoxplotData = eirKeyBoxplotData[eirKeyBoxplotData.Price < eirKeyBoxplotData.Price.quantile(0.75)*1.5]
```

```python
myColumns = []
myLabels = []

boxplot_pcs = eirKeyBoxplotData.groupby('eirKeyName')
for a, b in boxplot_pcs:
    myLabels.append(a)
    myColumns.append(b['Price'])
myLabels.insert(0, '') # label alignment hack - h/t McGyver :(
print len(myColumns)
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.boxplot(myColumns, myLabels)
ax.set_xticks(range(len(myLabels)))
ax.set_xticklabels(myLabels, rotation=90)
ax.set_xlabel('EirKeyName')
ax.set_ylabel('Price')
ax.set_title('Price v Post Code, 2016-to-Date')
plt.grid()
```

#### Observation
The boxplot is the most eloquent of all the graphs that show data, because it shows the most information in the least amount of space. Prices are too varied to sum up in a single point value.

### When Were the Houses Sold?
As noted above, it is remarkable that so many houses are being sold in Dublin 15. If we look at house sales broken down by date and by postal area, can we see if there is any pattern there?

```python
from bokeh.models import DatetimeTickFormatter, HoverTool

bp.output_notebook()

timeplot = bp.figure(width = 800, height = 494)

colors = ['red', 'purple', 'green', 'blue', 'black', 'coral', 'darkgreen']
markers = ['solid', 'dashed', 'dashdot']

color_counter = 0
marker_counter = 0
current_source = {}
renderer = {}
for pc, details in pcs:
    plot_from_zero = pd.DataFrame([0, 0])
    temp = details.copy()
    temp = temp[['Date', 'Address']]
    temp = temp.set_index('Date')
    temp = temp.resample('W').count()
    
    current_source = bp.ColumnDataSource(data = {"pc":[pc],
                                                    "sales":temp.Address.values})
    
    renderer[pc] = timeplot.line(temp.index,
                                 temp.Address.cumsum(),
                                 legend = pc,
                                 color = colors[color_counter],
                                 line_dash = markers[marker_counter],
                                 alpha = 0.25,
                                 line_width = 2,
                                 hover_alpha = 1,
                                hover_color = colors[color_counter])
    
    myTooltips = """
                <div>
                    <span style="font-size: 17px; font-weight: bold; color: {}">{}</span>
                </div>
                """
    
    timeplot.add_tools(HoverTool(tooltips = myTooltips.format(colors[color_counter], pc),
                                renderers = [renderer[pc]],
                                mode = 'mouse'))
    
    
    if color_counter == 6:
        color_counter = 0
    else:
        color_counter += 1
        
    if marker_counter == 2:
        marker_counter = 0
    else:
        marker_counter += 1
        
        
timeplot.title.text = "Count of Sales over Time (Hover over a plotline to see the eirKeyName)"
timeplot.xaxis.formatter = DatetimeTickFormatter()
timeplot.legend.location = (800, 400)
bp.show(timeplot)
```

#### Observation
There is no bump visible in the graph of Clonee/Dublin 15's expansion. The sales were steady all through the year - as, indeed, we can say for nearly all the districts. There is a jump visible for Killiney around March, and both the city centre districts, Dublin 1 and Dublin 2, tailed off after a fast start, but there is no seasonality per se visible in the chart.
